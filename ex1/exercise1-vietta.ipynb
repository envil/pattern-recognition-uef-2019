## Load data file `bogus_student_data.txt`

```python
import numpy as np
import pandas as pd

STUDENT_DATA_PATH = './bogus_student_data.txt'
data = np.genfromtxt(STUDENT_DATA_PATH, names=True)
```

## Visualize the data

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(data['exercise_points'])
plt.title('Exercises Points')
plt.show()
plt.hist(data['exam_points'])
plt.title('Exam Points')
plt.show()
plt.hist(data['grades'])
plt.title('Final Grade')
plt.show()
```

- We can't really infer so much from the histograms of the data other than the individual elements' distributions.

```python
plt.scatter(data['exercise_points'], data['grades'])
plt.title('Exercise//Grade')
plt.xlabel('Exercise Point')
plt.ylabel('Grade')
plt.show()
plt.scatter(data['exam_points'], data['grades'])
plt.title('Exam//Grade')
plt.xlabel('Exam Point')
plt.ylabel('Grade Point')
plt.show()
plt.scatter(data['exercise_points'], data['exam_points'])
plt.title('Exercise//Exam')
plt.xlabel('Exercise Point')
plt.ylabel('Exam Point')
plt.show()
```

We can now see some patterns here:
- The exercise points and grade points are proportional, i.e. the higher the exercise points are, the higher the grade points will be.
- The same themes apply to the relations between exam points and grade points, and exercise points and exam points
